Space: O(n) for the normalized string (strings are immutable, so each operation creates a new string).

Robustness:

Does not validate inputs (e.g., type checking for non-string inputs).

Method Choice Discussion (find vs. index):
.find(): Returns -1 if substring not found. No exception handling needed. Better for "safe search" as specified.
.index(): Raises ValueError if substring not found. Requires try-except block unless you're certain the substring exists.
Choice: .find() is more appropriate here because the spec explicitly asks for returning -1, avoiding exception overhead.

def normalize_and_find(text: str, needle: str) -> int:
    trimmed = text.strip()
    lowered = trimmed.lower()
    needle_lower = needle.lower()
    idx = lowered.find(needle_lower)
    return idx

```python
def normalize_and_find(text: str, needle: str) -> int:
    trimmed = text.strip()
    lowered = trimmed.lower()
    needle_lower = needle.lower()
    idx = lowered.find(needle_lower)
    return idx

print(normalize_and_find("  Hello World  ", "world"))
print(normalize_and_find("  Hello World  ", "WORLD"))
print(normalize_and_find("  Hello World  ", "foo"))
print(normalize_and_find("  foo  ", ""))
```
