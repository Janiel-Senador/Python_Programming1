```python
def cel_to_far(c: float) -> float:
    return (c * 9 / 5) + 32

def far_to_cel(f: float) -> float:
    return (f - 32) * 5 / 9

def main() -> None:
    print("xX Temperature Converter Xx")
    print("1 = Celsius to Fahrenheit")
    print("2 = Fahrenheit to Celsius")
    choice = input("Choose 1 or 2: ")
    if choice not in ("1", "2"):
        print("Invalid selection. Please restart and choose 1 or 2.")
        return

    value = input("Enter a temperature: ")
    try:
        temp = float(value)
    except ValueError:
        print("Invalid number. Please restart and enter a numeric temperature.")
        return

    if choice == "1":
        print(f"{temp}°C = {cel_to_far(temp)}°F")
    else:
        print(f"{temp}°F = {far_to_cel(temp)}°C")

if __name__ == "__main__":
    main()
```
