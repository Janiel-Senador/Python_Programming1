import random

def roll() -> int:
    return random.randint(1, 6)

def simulate_dice_sum(target_sum: int, trials: int) -> float:
    count = 0
    for _ in range(trials):
        die1 = roll()
        die2 = roll()
        if die1 + die2 == target_sum:
            count += 1
    return count / trials

# Run simulation
trials = 100000
empirical_prob = simulate_dice_sum(7, trials)
theoretical_prob = 6 / 36  # 6 ways to make 7 out of 36 outcomes

print(f"Trials: {trials}")
print(f"Empirical probability: {empirical_prob:.4f}")
print(f"Theoretical probability: {theoretical_prob:.4f}")
print(f"Difference: {abs(empirical_prob - theoretical_prob):.4f}")

# With seeding for reproducibility
random.seed(42)
empirical_seeded = simulate_dice_sum(7, trials)
print(f"\nWith seed(42): {empirical_seeded:.4f}")

```python
"""
Monte Carlo dice simulation to estimate the probability that
the sum of two fair 6-sided dice equals 7.

The Monte Carlo method uses repeated random sampling to approximate
a true probability. As the number of trials increases, the empirical
result should approach the theoretical probability.
"""

import random
from typing import List

def roll() -> int:
    return random.randint(1, 6)

def simulate_two_dice_rolls(num_rolls: int) -> float:
    count_sum_7 = 0
    for _ in range(num_rolls):
        total = roll() + roll()
        if total == 7:
            count_sum_7 += 1
    return count_sum_7 / num_rolls

def main() -> None:
    random.seed(42)
    num_trials = 100_000
    empirical_prob = simulate_two_dice_rolls(num_trials)
    theoretical_prob = 1 / 6
    print(f"Number of simulations: {num_trials}")
    print(f"Empirical P(sum = 7): {empirical_prob:.5f}")
    print(f"Theoretical P(sum = 7): {theoretical_prob:.5f}")
    print(f"Difference: {abs(empirical_prob - theoretical_prob):.5f}")

if __name__ == "__main__":
    main()
```
