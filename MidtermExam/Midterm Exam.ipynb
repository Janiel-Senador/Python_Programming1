(iii)Robustness: Works correctly for empty strings and when needle is not found. Does not validate inputs (e.g., type checking for non-string inputs). (vi) Improved Code:

```python
def normalize_and_find(text: str, needle: str) -> int:
    trimmed = text.strip()
    lowered = trimmed.lower()
    needle_lower = needle.lower()
    idx = lowered.find(needle_lower)
    return idx

print(normalize_and_find("  Hello World  ", "world"))
print(normalize_and_find("  Hello World  ", "WORLD"))
print(normalize_and_find("  Hello World  ", "foo"))
print(normalize_and_find("  foo  ", ""))
```

(b) ChatGPT Code: 

```python
def translate(text: str, mapping: dict[str, str]) -> str:
    """
    Translate text using a character mapping dictionary.
    
    Args:
        text: Input string to translate
        mapping: Dictionary mapping characters to replacements
    
    Returns:
        Translated string
    """
    text_lower = text.lower()
    result = []
    for char in text_lower:
        result.append(mapping.get(char, char))
    return ''.join(result)

# Test
leetspeak = {'a': '4', 'e': '3', 'i': '1', 'o': '0', 's': '5'}
print(translate("Hello World", leetspeak))  # h3ll0 w0rld
print(translate("Leet Speak!", leetspeak))  # l33t 5p34k!
```

```python
"""
Monte Carlo dice simulation to estimate the probability that
the sum of two fair 6-sided dice equals 7.

The Monte Carlo method uses repeated random sampling to approximate
a true probability. As the number of trials increases, the empirical
result should approach the theoretical probability.
"""

import random
from typing import List

def roll() -> int:
    return random.randint(1, 6)

def simulate_two_dice_rolls(num_rolls: int) -> float:
    count_sum_7 = 0
    for _ in range(num_rolls):
        total = roll() + roll()
        if total == 7:
            count_sum_7 += 1
    return count_sum_7 / num_rolls

def main() -> None:
    random.seed(42)
    num_trials = 100_000
    empirical_prob = simulate_two_dice_rolls(num_trials)
    theoretical_prob = 1 / 6
    print(f"Number of simulations: {num_trials}")
    print(f"Empirical P(sum = 7): {empirical_prob:.5f}")
    print(f"Theoretical P(sum = 7): {theoretical_prob:.5f}")
    print(f"Difference: {abs(empirical_prob - theoretical_prob):.5f}")

if __name__ == "__main__":
    main()
```

```python
"""
Vehicle system demonstrating inheritance and code reuse.

Classes:
- Vehicle (base class)
- Car (inherits Vehicle)
- Truck (inherits Vehicle and overrides drive behavior)

Inheritance avoids repeating common code such as attributes and methods.
"""
class Vehicle:
    def __init__(self, color: str, mileage: float, fuel_liters: float) -> None:
        self.color = color
        self.mileage = mileage
        self.fuel_liters = fuel_liters

    def drive(self, km: float, km_per_liter: float) -> None:
        fuel_needed = km / km_per_liter
        fuel_used = min(fuel_needed, self.fuel_liters)
        self.fuel_liters -= fuel_used
        self.mileage += fuel_used * km_per_liter

    def refuel(self, liters: float) -> None:
        self.fuel_liters += liters

    def __str__(self) -> str:
        return (f"{self.__class__.__name__}(color={self.color}, "
                f"mileage={self.mileage:.1f}, fuel_liters={self.fuel_liters:.1f})")

class Car(Vehicle):
    pass

class Truck(Vehicle):
    def drive(self, km: float, km_per_liter: float) -> None:
        adjusted_efficiency = km_per_liter * 0.8
        super().drive(km, adjusted_efficiency)


car = Car(color="Red", mileage=10000, fuel_liters=40)
truck = Truck(color="Blue", mileage=50000, fuel_liters=80)

print("Initial state:")
print(car)
print(truck)

car.drive(100, km_per_liter=10)
truck.drive(100, km_per_liter=10)

print("\nAfter driving 100 km:")
print(car)
print(truck)

car.refuel(5)
truck.refuel(10)

print("\nAfter refueling:")
print(car)
print(truck)
```

```python
class Dog:
    def __init__(self, name: str, age: int, breed: str) -> None:
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self) -> None:
        print(f"{self.name} says: Woof!")

class Kennel:
    def __init__(self) -> None:
        self.dogs = []

    def add_dog(self, dog: Dog) -> None:
        self.dogs.append(dog)

    def remove_dog(self, name: str) -> None:
        self.dogs = [dog for dog in self.dogs if dog.name != name]

    def oldest_dog(self) -> Dog:
        if not self.dogs:
            return None
        oldest = self.dogs[0]
        for dog in self.dogs:
            if dog.age > oldest.age:
                oldest = dog
        return oldest

    def find_by_breed(self, breed: str) -> list:
        result = []
        for dog in self.dogs:
            if dog.breed == breed:
                result.append(dog)
        return result

dog1 = Dog("Buddy", 5, "Labrador")
dog2 = Dog("Max", 3, "Beagle")
dog3 = Dog("Rocky", 7, "Labrador")
kennel = Kennel()
kennel.add_dog(dog1)
kennel.add_dog(dog2)
kennel.add_dog(dog3)
dog1.bark()
print("Oldest dog:", kennel.oldest_dog().name)
print("Labradors:", [dog.name for dog in kennel.find_by_breed("Labrador")])
kennel.remove_dog("Max")
print("Remaining dogs:", [dog.name for dog in kennel.dogs])
```
